```
from scipy import stats
```

```
import numpy as np
```

#One Sample T-test

```
rvs = stats.norm.rvs(loc = 5, scale = 2, size = 50)
rvs
```

```
stats.ttest_1samp(rvs,5.0)
```

```
stats.ttest_1samp(rvs, 0.0)
```

#2-sample T-Test

```
rvs1 = stats.norm.rvs(loc =5.0, scale = 2.0, size = 500)
rvs2 = stats.norm.rvs(loc = 5.0, scale = 1.0, size = 450)
```

```
stats.ttest_ind(rvs1,rvs2, equal_var = False)
```

#F-Test 
determines if the variance of the population is the same given the same mean

```
stats.f_oneway(rvs, rvs1, rvs2)
```

#Analysis of variance (ANOVA)
is a statistical procedure used to analyze whether the variance of a population is caused by the differences between several (K>2) pre-defined sub-populations, or groups.

```
stats.f_oneway(rvs, rvs1, rvs2)
```
